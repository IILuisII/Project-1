```python
super_df = pd.read_csv('superbowl.csv',low_memory=False)
super_df

super_df['Winner'].value_counts()

#for index, row in super_df.iterrows():
#print()
```

```python
x = pd.DataFrame(super_df['Winner'].value_counts())

x.reset_index(inplace = True)
x.columns = ['team','winner']
plt.figure(figsize=(16, 6))

plt.bar(x.team, x.winner)
plt.xticks(rotation=65, horizontalalignment='right')

plt.ylabel('Number of Wins')
plt.xlabel('Teams')
plt.title('Superbowl Winners 1967 - 2020')

plt.show()
```

```python
super_df
super_df[['Month','Day','Year']] = super_df.Date.str.split(expand=True) 

winners =super_df[['Year','Winner','Winner Pts','Date']]
losers =super_df[['Year','Loser','Loser Pts','Date']]
```

```python
x.sort_values(by=['winner'], inplace=True,ascending=False)
x.iloc[0].team
```

```python

# Data to plot
labels =[ x.iloc[0].team, x.iloc[1].team, x.iloc[2].team, x.iloc[3].team, x.iloc[4].team]
sizes =[ x.iloc[0].winner, x.iloc[1].winner, x.iloc[2].winner, x.iloc[3].winner, x.iloc[4].winner]

#sizes = [ x[0].winner, x[1].winner, x[2].winner, x[3].winner, x[4].winner]
colors = ['blue', 'gold', 'lightcoral', 'lightskyblue','yellowgreen']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False, startangle=140)
plt.title('Top 5 Winningest Teams')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(16, 6))

plt.bar(winners['Year'],winners['Winner Pts'])
plt.bar(losers['Year'],losers['Loser Pts'])
plt.xticks(rotation=65, horizontalalignment='right')
plt.title('Superbowl Margins of Victory 1967-2020')
#plt.legend()
#plt.legend(['Winner Pts', 'Loser Pts'], ['Winner Pts','Loser Pts'])
orange_patch = mpatches.Patch(color='coral', label='Loser')
blue_patch = mpatches.Patch(color='royalblue', label='Winner')
plt.legend(handles=[orange_patch,blue_patch])

plt.ylabel('Scores')
plt.xlabel('Years')
plt.show()
```
